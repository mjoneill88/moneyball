```python
all_data_no_na = all_data.copy()
```

```python
all_data_no_na = all_data_no_na.dropna()
len(all_data_no_na)
```

```python
all_data_no_na = all_data_no_na.drop_duplicates()
len(all_data_no_na)
```

```python
#OBP Formula, new column
all_data_no_na['OBP'] = (all_data_no_na['H']+all_data_no_na['BB']+all_data_no_na['HBP'])/(all_data_no_na['AB']+all_data_no_na['BB']+all_data_no_na['HBP']+all_data_no_na['SF'])
```

```python
all_data_no_na.head()
```

```python
all_data_no_na['Ratio'] = all_data_no_na['OBP']/all_data_no_na['salary']
```

```python
moneyball = all_data_no_na[all_data_no_na['AB']>30]
```

```python
moneyball.sort_values('Ratio', ascending=False).head()
```
